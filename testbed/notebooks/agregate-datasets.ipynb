```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
from ipywidgets import *
import warnings
warnings.simplefilter(action='ignore', category=Warning)
%matplotlib inline
```

```python
columnz = ['Trace_no', 'Mode', 'Total_host', 'Host_no', 'Segment_no', 
           'Arr_time', 'Del_Time', 'Stall_Dur', 'Rep_Level', ' Del_Rate', 'Act_Rate',
           'Byte_Size', 'Buff_Level', 'Algorithm', 'Seg_Dur', 'Width', 'Height',
           'Play_Pos', 'RTT_app', 'P1203', 'Clae', 'Duanmu', 'Yin', 'Yu',
           'persegment_RTT', 'Throughput', 'Packets']
```

```python
datasets = {
    "normal":      "../Data/Processed/dataset_complete_(size:3600x27)_(tag:normal).csv",
    "duplicate5":  "../Data/Processed/dataset_complete_(size:3600x27)_(tag:duplicate5).csv",  
    "duplicate30": "../Data/Processed/dataset_complete_(size:3600x27)_(tag:duplicate30).csv",  
    "duplicate60":  "../Data/Processed/dataset_complete_(size:3600x27)_(tag:duplicate60).csv",     
    "reorder5":    "../Data/Processed/dataset_complete_(size:3600x27)_(tag:reorder5).csv",
    "reorder30":   "../Data/Processed/dataset_complete_(size:3600x27)_(tag:reorder30).csv",
    "reorder60":   "../Data/Processed/dataset_complete_(size:3600x27)_(tag:reorder60).csv",    
    "corrupt5":    "../Data/Processed/dataset_complete_(size:3600x27)_(tag:corrupt5).csv",   
    "corrupt30":   "../Data/Processed/dataset_complete_(size:3600x27)_(tag:corrupt30).csv",       
    "corrupt60":   "../Data/Processed/dataset_complete_(size:3600x27)_(tag:corrupt60).csv",           
}
megaDf = pd.DataFrame(columns = columnz)
for tag, file in datasets.items():
    df = pd.read_csv(file, names = columnz, skiprows=[0])
    df["tag"] = tag
    print ("tag= ", tag, " ==> shape: ", df.shape)
    megaDf = pd.concat([megaDf, df])    
print ("Final shape: ", megaDf.shape)    
```
